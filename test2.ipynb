```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/murpi/wilddata/master/quests/cars.csv", sep=',')

df.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize = (12, 12))
fig.suptitle("Analyses Voitures")

count_df = df['continent'].value_counts().reset_index()
count_df.columns = ['continent', 'count']

chart_1 = plt.subplot(221)
sns.barplot(count_df,
                x = "continent",
                y = "count",
                hue = "continent")


chart_2 = plt.subplot(222)
sns.barplot(df,
                x = "continent",
                y = "weightlbs",
                hue = "continent")


chart_3 = plt.subplot(223)
sns.scatterplot(count_df,
                x = "continent",
                y = "count",
                hue = "continent")


chart_4 = plt.subplot(224)
sns.barplot(count_df,
                x = "continent",
                y = "count",
                hue = "continent")


plt.show()
```
